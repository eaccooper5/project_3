```python
import csv
from sqlalchemy import create_engine
import pandas as pd
from IPython.display import display
```

```python
# Create an engine
engine = create_engine("postgresql://postgres@127.0.0.1:5432/project_3")
```

```python
# Create a connection
connection = engine.connect()
```

```python
# Execute a query
result = connection.execute("SELECT * FROM weo_data_emerging_5regions")
rows = result.fetchall()
```

```python
# Print the results
for row in rows:
    print(row)
```

```python
# Create a string by joining each tuple's string representation with a newline character
string_data = '\n'.join(map(str, rows))
```

```python
# Split the string into lines
lines = string_data.split('\n')
```

```python
# Filter out empty lines and lines starting with comments
filtered_lines = [line for line in lines if line.strip() and not line.strip().startswith('#')]
```

```python
# Print the resulting lines
for line in filtered_lines:
    print(line)
```
